```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
boston = datasets.load_boston()

X = boston.data
y = boston.target

X = X[y < 50.0]
y = y[y < 50.0]
```

```python
from playML.LinearRegression import LinearRegression
```

```python
reg = LinearRegression()
reg.fit_normal(X_train, y_train)
```
